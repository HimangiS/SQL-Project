```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv(r'C:\DS\SQL\Project\My Project/top_batsmen.csv')

# Display the DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(df['batsman'], df['strike_rate'], color='skyblue')
plt.xlabel('Strike Rate')
plt.ylabel('Batsmen')
plt.title('Top Batsmen by Strike Rate')
plt.gca().invert_yaxis()  # Invert y-axis to display highest strike rate at the top
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define colors for the bars
colors = ['skyblue', 'salmon', 'lightgreen', 'gold', 'lightcoral', 
          'lightseagreen', 'orchid', 'cornflowerblue', 'mediumaquamarine', 'khaki']

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['bowler'], df['total_wickets'], color=colors)
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.title('Top Wicket-Takers in IPL')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
city_data = {
    'city': ['Jaipur', 'Abu Dhabi', 'Visakhapatnam', 'Kochi', 'Dharamsala', 'Port Elizabeth', 'Hyderabad', 'Indore', 'Bangalore', 'Mumbai', 'Delhi', 'Pune', 'East London', 'Durban', 'Sharjah', 'Cape Town', 'NA', 'Bengaluru', 'Raipur', 'Nagpur', 'Kimberley', 'Centurion', 'Kolkata', 'Ahmedabad', 'Chandigarh', 'Johannesburg', 'Bloemfontein', 'Kanpur', 'Cuttack', 'Ranchi', 'Rajkot', 'Chennai', 'Dubai'],
    'matches_count': [47, 29, 13, 5, 9, 7, 64, 9, 65, 101, 74, 38, 3, 15, 12, 7, 13, 15, 6, 3, 3, 12, 77, 12, 56, 8, 2, 4, 7, 7, 10, 57, 26]
}

# Create DataFrame
df = pd.DataFrame(city_data)

# Sort DataFrame by matches_count in descending order
df = df.sort_values(by='matches_count', ascending=False)

# Custom colors
colors = ['skyblue' if x < max(df['matches_count']) else 'orange' for x in df['matches_count']]

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(df['city'], df['matches_count'], color=colors)
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.title('IPL Matches Held in Each City')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
